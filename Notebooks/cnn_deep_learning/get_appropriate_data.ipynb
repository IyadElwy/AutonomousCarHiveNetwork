```python
import pandas as pd
import numpy as np
import os
```

```python
dataset_l_r = []  # f, b, r, l
dataset_l_r_none = []

for test_dir in os.listdir('../../data/sharpened__only_lane_region'):
    df = pd.read_csv(f'../../data/sharpened__only_lane_region/{test_dir}/steering.csv')
    df['4'] = df['4'].map(lambda x: x.split('../data/')[1])
    dataset_l_r.extend(df.loc[:, ['2', '3', '4']][((df['2'] == 0) & (df['3'] == 1)) |
                                                  ((df['2'] == 1) & (df['3'] ==0))]
                       .values.tolist())
    dataset_l_r_none.extend(df.loc[:, ['2', '3', '4']][((df['2'] == 0) & (df['3'] == 1)) |
                                                       ((df['2'] == 1) & (df['3'] ==0)) |
                                                       ((df['2'] == 0) & (df['3'] ==0))
                            ]

                            .values.tolist())
```

```python
dataset_l_r_none
```

```python
dataset_l_r = pd.DataFrame(dataset_l_r, columns=['r', 'l', 'file'])
dataset_l_r_none = pd.DataFrame(dataset_l_r_none, columns=['r', 'l', 'file'])
```

```python
dataset_l_r.to_csv('../../data/sharpened__only_lane_region/dataset_l_r.csv', index=False)
```

```python
dataset_l_r_none.to_csv('../../data/sharpened__only_lane_region/dataset_l_r_none.csv', index=False)
```
